```python
%matplotlib inline

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time
import json
```

```python
fname = 'sp-pos-quot-dep-2021-03-28-19h20-reg.csv'
d = pd.read_csv(fname, 
                sep = ';', 
                parse_dates = ['jour'], 
                dtype = {'P': int, 'T': int},
                skiprows = [1])
d.sample(3)
```

```python
d.dtypes
```

```python
regs = d['reg'].unique()
ages = d['cl_age90'].unique()
```

```python
three_class = {0: 'whole',
               9: '0-19',
               19: '0-19',
               29: '20-59',
               39: '20-59',
               49: '20-59',
               59: '20-59',
               69: '60+',
               79: '60+',
               89: '60+',
               90: '60+',
              }
```

```python
d['three_class'] = d['cl_age90'].map(lambda x: three_class[x])
d.sample(3)
```

```python
dg = d.groupby(['reg', 'jour', 'three_class']).agg([sum])
dg.sample(3)
```

```python
dg = (dg.reset_index()
      .drop(columns = ['cl_age90'])
            )
dg.sample(3)
```

```python
dg.tail(3)
```

```python
fname = 'sp-pos-quot-dep-2021-03-28-19h20.csv'
fname = fname[:-4] + '-reg3C.csv'
dg.to_csv(fname, sep = ';', index = False)
```
