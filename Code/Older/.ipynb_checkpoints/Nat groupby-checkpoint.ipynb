```python
%matplotlib inline

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time
import json
```

```python
fname = 'sp-pos-quot-dep-2021-03-28-19h20-reg.csv'
d = pd.read_csv(fname, 
                sep = ';', 
                parse_dates = ['jour'], 
                dtype = {'P': int, 'T': int},
                skiprows = [1])
d.sample(3)
```

```python
dg = (d.groupby(['jour', 'cl_age90'])
      .agg([sum])
      .drop(columns = ['reg'])
      .reset_index()
     )
dg
```

```python
fname = 'sp-pos-quot-dep-2021-03-28-19h20.csv'
fname = fname[:-4] + '-nat.csv'
dg.to_csv(fname, sep = ';', index = False)
```
