```python
fname_temp = '../Temp/sp-pos-quot-dep-2021-04-03-19h33'
fname = fname_temp + '-tot-3C.csv'
d = pd.read_csv(fname, 
                sep = ';', 
                parse_dates = ['jour'], 
                dtype = {'P': int, 'T': int},
                skiprows = [1])
d.tail(3)
```

```python
d = d[ d['jour'] == dt.datetime(2021, 3, 29)]
```

```python
regs = d['entity'].unique()
three_class = d['three_class'].unique()
```

```python
reg_3C_pop = {}
for reg in regs:
    reg_3C_pop[reg] = {}
    for _class in three_class:
        reg_3C_pop[reg][_class] = int(d
                                   [(d['entity'] == reg)
                                   & (d['three_class'] == _class)
                                    ]
                                   ['pop'])
reg_3C_pop
```

```python
dg = d.groupby(['three_class']).agg([sum])
dg
```

```python
dg = (dg.reset_index()
      .drop(columns = ['entity'])
            )
dg
```

```python
fname = 'reg_3C_pop.json'
with open(fname, 'w') as outfile:
    outfile.write(json.dumps(reg_3C_pop))
```
